```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_base0 = pd.read_csv('./submissions/aguiar_feature_matrix_5cv_sub_789.csv')
df_base1 = pd.read_csv('./submissions/kageyama_90_sub_794.csv')
df_base2 = pd.read_csv('./submissions/kageyama_92_sub.csv')
df_base3 = pd.read_csv('./submissions/kageyama_94_sub_793.csv')
df_base4 = pd.read_csv('./submissions/kageyama_96_sub.csv') 
df_base5 = pd.read_csv('./submissions/kageyama_98_sub.csv')
df_base6 =  pd.read_csv('./submissions/kageyama_feature_matrix_25cv_tuned_by_Ivan_noIDs_delmissing_sub_794.csv') # 0.2229
df_base7 =  pd.read_csv('./submissions/kageyama_feature_matrix_25cv_tuned_by_Ivan_sub_noIDs_795.csv') # 0.2233
df_base8 =  pd.read_csv('./submissions/publics/lgb157_0.7911.csv')
df_base9 =  pd.read_csv('./submissions/publics/sub_798.csv')
df_base10 = pd.read_csv('./submissions/publics/tidy_xgb_0.78847.csv')
df_base11 = pd.read_csv('./submissions/publics/lgb715_0.7906.csv')
df_base12 = pd.read_csv('./submissions/publics/lightGBM_test_predictions_rank_mean.csv')
```

```python
df_base0.head()
```

```python
#CREATING SOLUTIONS COLUMNS
df_base = pd.merge(df_base0,df_base1,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base2,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base3,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base4,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base5,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base6,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base7,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base8,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base9,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base10,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base11,how='inner',on='SK_ID_CURR')
df_base = pd.merge(df_base,df_base12,how='inner',on='SK_ID_CURR')
```

```python
#CORRELATION MATRIX (Pearson Correlation to measure how similar are 2 solutions)
plt.figure(figsize=(12,12))
sns.heatmap(df_base.iloc[:,1:].corr(),annot=True,fmt=".2f")
```
