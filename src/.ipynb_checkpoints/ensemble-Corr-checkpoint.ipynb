```python
df_base0 = pd.read_csv('./csv/base0_0.2211.csv')
df_base1 = pd.read_csv('./csv/base1_0.2212.csv')
df_base2 = pd.read_csv('./csv/base2_0.2212.csv')
df_base3 = pd.read_csv('./csv/lgsub_miyano_ver6.csv') # 0.2219
df_base4 = pd.read_csv('./csv/lgsub_miyano_ver5.csv') # 0.2219
df_base5 = pd.read_csv('./csv/lgsub_miyano_ver9.csv') # ???
df_base6 =  pd.read_csv('./csv/lgsub_miyano_ver8.csv') # 0.2229
df_base7 =  pd.read_csv('./csv/lgb_metatext_activefeat_TFIDF.csv') # 0.2233
df_base8 =  pd.read_csv('./csv/lgb_metatext_activefeat_TFIDF2.csv') # 0.2233
df_base9 =  pd.read_csv('./csv/lgb_metatext_activefeat_TFIDF3.csv') # 0.2233
df_base10 = pd.read_csv('./csv/base14_0.2237.csv')
df_base11 = pd.read_csv('./csv/base18_0.2246.csv')

```

```python
#CREATING SOLUTIONS COLUMNS
df_base = pd.merge(df_base0,df_base1,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base2,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base3,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base4,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base5,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base6,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base7,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base8,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base9,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base10,how='inner',on='item_id')
df_base = pd.merge(df_base,df_base11,how='inner',on='item_id')
```

```python
#CORRELATION MATRIX (Pearson Correlation to measure how similar are 2 solutions)
plt.figure(figsize=(12,12))
sns.heatmap(df_base.iloc[:,1:].corr(),annot=True,fmt=".2f")
```
